```python
df_authors = pd.read_csv('D:\Projects\BIG_DATA_TWITTER_ANALYSIS\Dataset\GraphConnections\\total_connections\\authors_l_w.csv')
df_authors
```

```python
def get_wcc_id(text):
    try:
        return df_wcc[df_wcc['nodeId'] == text]['propertyValue'].values[0]
    except:
        return 0
```

```python
def get_pagerank_id(text):
    try:
        return df_pagerank[df_pagerank['nodeId'] == text]['propertyValue'].values[0]
    except:
        return 0
```

```python
df_authors['pagerank'] = df_authors['Id'].apply(get_pagerank_id)
df_authors
```
